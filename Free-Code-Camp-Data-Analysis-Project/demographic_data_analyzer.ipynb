Task 5 - What is the minimum number of hours a person works per week?

```python
Data['hours-per-week'].min()
```

Task 6 - What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_hours = Data['hours-per-week'].min()

min_hours_workers = Data[Data['hours-per-week'] == min_hours]
high_earners = min_hours_workers[min_hours_workers['salary'] == '>50K']

percentage = (len(high_earners) / len(min_hours_workers)) * 100

print(f"The percentage of people who work the minimum number of hours per week and have a salary of more than 50K is {percentage:.2f}%.")
```

Task 7 - What country has the highest percentage of people that earn >50K and what is that percentage?

```python
big_money_earners = Data[Data['salary'] == '>50K']

high_earners_per_country = big_money_earners['native-country'].value_counts()
total_people_per_country = Data['native-country'].value_counts()

percentage_high_earners = (high_earners_per_country / total_people_per_country) * 100

highest_percentage_country = percentage_high_earners.idxmax()
highest_percentage = percentage_high_earners.max()

print(f"The country with the highest percentage of people earning >50K is {highest_percentage_country} with {highest_percentage:.2f}%.")
```

Task 8 - Identify the most popular occupation for those who earn >50K in India.

```python
india_high_earners = Data[(Data['salary'] == '>50K') & (Data['native-country'] == 'India')]
india_high_earners
```

```python
most_popular_job = india_high_earners['occupation'].max()
print(f'The most popular job in India is {most_popular_job}')
```
