```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
Data = pd.read_csv('epa-sea-level.csv')
Data
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('year')
plt.ylabel('Sea Level')
plt.scatter(Data['Year'], Data['CSIRO Adjusted Sea Level'])
```

```python
slope, intercept, r, p, se = linregress(Data['Year'], Data['CSIRO Adjusted Sea Level'])
```

```python
Forcast_Years = np.arange(Data['Year'].min(),2051, 1)  #  x value
```

```python
Forcast_Level = slope * Forcast_Years + intercept
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('year')
plt.ylabel('Sea Level')

plt.plot(Forcast_Years, Forcast_Level, color = 'red', label = 'Line')
plt.scatter(Data['Year'], Data['CSIRO Adjusted Sea Level'], label = 'Data Points')

plt.legend()
```

```python
Forcast_2_Data = Data[Data['Year'] >= 2000]
```

```python
slope_2, intercept_2, r_2, p_2, se_2 = linregress(Forcast_2_Data['Year'], Forcast_2_Data['CSIRO Adjusted Sea Level'])
```

```python
Forcast_Years_2 = np.arange(2000,2050,1)
```

```python
Forcast_Level_2 = slope_2 * Forcast_Years_2 + intercept_2
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

plt.plot(Forcast_Years, Forcast_Level, color = 'red', label = 'Overall Change in Avg Sea Level')
plt.plot(Forcast_Years_2, Forcast_Level_2, color = 'green', label = 'Recent Change in Avg Sea Level')
plt.scatter(Data['Year'], Data['CSIRO Adjusted Sea Level'], label = 'Sea Level by Years')

plt.legend()
```
